Contador de nulos por columna

```python
from pyspark.sql.functions import *

def nulos_columns(df):

    df.select([
        sum( when(col(column).isNull(),1).otherwise(0) ).alias(column)
        for column in df.columns
    ]).show()
```

Importacion de CSVs, Exploracion de datos y contador de nulos por columnas
