```python
#Importar las librerías necesarias para trabajar con Spark
from pyspark.sql import SparkSession 
spark = SparkSession.builder.appName("motorSpark").getOrCreate()
```

Funcion de exploracion de datos 

```python
def explorar_datos(df):
    print("\n-- Muestra --")
    df.show(5)
    
    print("\n-- Esquema --")
    df.printSchema()
    
    print(f"\nTotal de filas: {df.count()}")
    
    print("\n-- Estadísticas descriptivas --")
    df.describe().show(truncate=False)
```

```python
order_payments = spark.read.csv("olist_order_payments_dataset.csv", header=True, inferSchema=True)
explorar_datos(order_payments)
```

```python
costumers = spark.read.csv("olist_customers_dataset.csv", header=True, inferSchema=True)
explorar_datos(costumers)
```
