Explorador de dataframes

```python
def explorar_datos(df):
    print("\n-- Muestra --")
    df.show(5)
    print("\n-- Esquema --")
    df.printSchema()
    print(f"\nTotal de filas: {df.count()}")
    print("\n-- Estadísticas descriptivas --")
    df.describe().show(truncate=False)
```

Contador de nulos por columna

```python
from pyspark.sql.functions import *

def nulos_columns(df):
    print("nulos \n")
    df.select([
        sum( when(col(column).isNull(),1).otherwise(0) ).alias(f"nulos {column}")
        for column in df.columns
    ]).show()
```

Contador de duplicados 

```python
def contador_duplicados(df):
    print("Duplicados \n")
    df.groupBy(df.columns).count().filter("count > 1").show()
```

ventas de producto por mes

```python
ventas_producto = (order
    .filter(col("order_status") == "delivered")  # solo pedidos completados
    .join(order_items,  "order_id",    "left")
    .join(order_payments,     "order_id",    "left")
    .join(products,     "product_id",  "left")
    .select(
        # para saber cuándo se compró
        "order_id",
        "order_purchase_timestamp",
        
        # para saber qué se vendió
        "product_category_name",
        
        # para saber cuánto se vendió
        "payment_value",

        "order_status"
    )
)
ventas_producto.show(5)
```

```python
from pyspark.sql.functions import month, year, col

ventas_producto = (ventas_producto
    .withColumn("mes", month(col("order_purchase_timestamp")))
    .withColumn("año", year(col("order_purchase_timestamp")))
)
```

```python
from pyspark.sql.functions import sum, count, avg, round

ventas_trimestre = (ventas_producto
    .groupBy("año", "mes","product_category_name")
    .agg(
        round(sum("payment_value"),  2).alias("total_valor_ventas"),
        count("order_id").alias("total_pedidos")
    )
    .orderBy("año", "mes", "total_valor_ventas", ascending=False)
)

ventas_trimestre.show(90)
```


promedio de tiempo de entrega de pedidos

```python
tiempos = (order
    .filter(col("order_status") == "delivered")  
    .filter(col("order_delivered_customer_date").isNotNull())
)
tiempos.show(5)
```

```python
from pyspark.sql.functions import datediff, when, round, col

tiempos = (tiempos
    
    .withColumn("dias_entrega_real",
        datediff(col("order_delivered_customer_date"), col("order_purchase_timestamp"))
    )

    .withColumn("dias_retraso",
        datediff(col("order_delivered_customer_date"), col("order_estimated_delivery_date"))
    )

    .withColumn("dias_aprobacion",
        datediff(col("order_approved_at"), col("order_purchase_timestamp"))
    )

    .withColumn("estado_entrega",
        when(col("dias_retraso") < 0,  "entregó antes")
        .when(col("dias_retraso") == 0, "entregó a tiempo")
        .when(col("dias_retraso") <= 7, "retraso leve")
        .otherwise("retraso grave")
    )
)
```

```python
from pyspark.sql.functions import avg, count, sum

tiempos.agg(
    round(avg("dias_entrega_real"), 1).alias("promedio_dias_entrega"),
    round(avg("dias_retraso"), 1).alias("promedio_dias_retraso"),
    round(avg("dias_aprobacion"), 1).alias("promedio_dias_aprobacion")
).show()

# Distribución del estado de entrega
print("=== DISTRIBUCIÓN ENTREGAS ===")
tiempos.groupBy("estado_entrega") \
    .agg(
        count("order_id").alias("total_pedidos"),
        round(avg("dias_retraso"), 1).alias("promedio_retraso")
    ) \
    .orderBy("total_pedidos", ascending=False) \
    .show()
```
